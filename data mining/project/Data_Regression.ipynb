```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
pd.set_option("display.max_columns",100)
import warnings
warnings.simplefilter('ignore')
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('E:\data mining\dataset_csv\dataset_imp_date_time.csv')
data
```

```python
v=data.iloc[:,2:].corr()
```

```python
sns.heatmap(v)
```

```python
data1=data.iloc[:,4:]
```

```python
data['Battery_Replacement_date1']=pd.to_datetime(data['Battery_Replacement_date1']).values.astype(np.int64)// 10 ** 6
data['send_time1']=pd.to_datetime(data['send_time1']).values.astype(np.int64)// 10 ** 6
```

```python
X=data.drop(['CURRENT_YEARLY_MILEAGE'],axis=1)
y=data['CURRENT_YEARLY_MILEAGE']
X=X.iloc[:,4:]
X
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
x_train.shape
```

```python
x_test.shape
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression,SGDRegressor,ElasticNet,Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import KFold,cross_val_score
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn import linear_model
from sklearn import metrics

regr = linear_model.LinearRegression()
regr.fit(x_train, y_train)
y_scores = regr.predict(x_test)
scores = cross_val_score(regr.fit(x_train, y_train), x_train, y_train, cv=10)
MSE = metrics.mean_squared_error(y_test, y_scores)
SD = scores.std()
print(MSE)
print(SD)
```

```python

regr1 = KNeighborsRegressor(n_neighbors = 10)
regr1.fit(x_train, y_train)
y_scores = regr1.predict(x_test)
scores = cross_val_score(regr1.fit(x_train, y_train), x_train, y_train, cv=10)
MSE = metrics.mean_squared_error(y_test, y_scores)
SD = scores.std()
print(MSE)
print(SD)
```

```python
from sklearn.linear_model import ElasticNet

en = ElasticNet()
en.fit(x_train, y_train)
y_scores = en.predict(x_test)
scores = cross_val_score(en.fit(x_train, y_train), x_train, y_train, cv=10)
MSE = metrics.mean_squared_error(y_test, y_scores)
SD = scores.std()
print(MSE)
print(SD)
```

```python
from sklearn.linear_model import Lasso

clf = linear_model.Lasso()
clf.fit(x_train, y_train)
y_scores = clf.predict(x_test)
scores = cross_val_score(clf.fit(x_train, y_train), x_train, y_train, cv=10)
MSE = metrics.mean_squared_error(y_test, y_scores)
SD = scores.std()
print(MSE)
print(SD)
```

```python

models=[]
models.append(("linear regression",LinearRegression()))
#models.append(("SGDRegressor",SGDRegressor()))
models.append(("KNeighborsRegressor",KNeighborsRegressor()))
models.append(("Elastic",ElasticNet()))
models.append(("Lasso",Lasso()))
#models.append(("DecisionTreeRegressor",DecisionTreeRegressor()))
#models.append(("Polynomial_regression",LinearRegression()))
results=[]
names=[]
for name,model in models:
    kfold=KFold(n_splits=5,random_state=10)
    cv_resilt=cross_val_score(model,x_train,y_train,cv=kfold,scoring="neg_mean_squared_error")
    results.append(cv_resilt)
    names.append(name)
    msg="%s:%f(%f)"%(name,cv_resilt.mean(),cv_resilt.std())
    print(msg)
    #print(cv_resilt)
```
